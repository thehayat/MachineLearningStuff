### Data Visualisation for beginners:###
Few exploration around the the matplotlib. A very useful library used to plot data. Analysts and Machine learning engineers use this extensively. 
This notebook contains basic usage of matplotlib. 

```python
"Importing Libraries"
import matplotlib.pyplot as plt
%matplotlib inline

print("Done Importing")
```

```python
# Reading the data file

with open ("Goals.txt",'r') as f:
    
    homeTeamGoals = [x for x in f.readline().strip("\n").strip(" ").split(" ")]
    awayTeamGoals = [x for x in f.readline().strip("\n").strip(" ").split(" ")]

"About the data"
# The data contains two list of goals scored by two teams. Let's say "HomeTeamGoals" and  "AwayTeamGoals" in the certain 
#  number of matches between then. Let's do the exploration and find out which team turn out to best.
```

```python


x_axis = [i for i in range(len(homeTeamGoals))] #number of matachs between the teams. Let's make it x-axis
print("Number of matches: ",len(x_axis))

plt.figure(figsize=(10,5)) # Customly define the figure of the plot. Figsize=(l * b)

plt.plot(homeTeamGoals , ls="--",label="homeTeamGoals") # plot a simple curve graph for homeTeamGoals.
plt.plot(awayTeamGoals, ls= ":",label="awayTeamGoals") # plot a simple curve graph for homeTeamGoals.

plt.xlabel("Game number") # Define the label of x-axis.
plt.ylabel("Goals Scored",rotation=90) # Define the label of x-axis. Rotation flips the label appearence

plt.legend(loc="upper right") #plot the legend on the graph. location of the legend can be defined by 'loc' 

plt.title(r"Our first graph $\frac{1}{2}$") # Lot of cutomisation can be done using r and we can write that in title.
plt.xlim(-1,160) # We can set the limit of the x-axis manually here.
plt.ylim(0,7) # We can set the limit of the y-axis manually here.

plt.text(50,6,"My custom text",fontsize=10,c="b") # With the this you can specify some text to your plot.
plt.savefig("myfig.pdf") #Save fig locally with format like .png , .pdf , .jpg 
plt.show() # Shows the figure to console.

"Note: After show method is called now everything will correspond the new plot. Because previous plot is executed already"

# Scatter plot
plt.scatter(x_axis,homeTeamGoals, c='b') #Call scatter method and pass x and y.
plt.xlabel(r"Game number$_5$") 
plt.xticks([0,50,100],["zero","fifty","hundred"]) # With ticks we can replace the interval data such as [0,10,20,30,40] to [zero,ten,twenty,thrity,forty]
plt.show() #everything after the show() correspond to a new plot
plt.scatter(x_axis,awayTeamGoals, c='r')
```

```python
# plt.hist(homeTeamGoals,bins=range(10),label="home team goal") # pasing single list in x. If we want to pass multiple data.

plt.hist(x=(homeTeamGoals,awayTeamGoals),bins=range(10),label=["Away team goal","Away Team Goal"]) # We pass the tuple containing the multiple 
plt.legend()
plt.xlabel("goals scored")
plt.ylabel("Number of goals scored in matches")
plt.show()
```
